```python
def sigmoid(x):
    '''Common activation function, 'squeezes' output to range [0, 1]'''
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def mse(a, y):
        return np.mean((a - y)**2)
    
def mse_derivative(a, y):
    return 2 * np.sum(a - y)/a.shape[0]
```

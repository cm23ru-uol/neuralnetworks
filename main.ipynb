```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
class NeuralNetwork:
    
    def __init__(self):
        self.size = [2, 4, 4, 2]
        self.weights = [np.random.random((a, b)) for a,b in zip(self.size[1:], self.size[:-1])]
        self.biases = [np.random.random(i).reshape((-1, 1)) for i in self.size[1:]]
        
    def feedforward(self, a):
        for weights, biases in zip(self.weights, self.biases):
            a = sigmoid(np.matmul(weights, a) + biases)
            
        return a 
```

```python
nn = NeuralNetwork()

a = np.array([[10], [10]])
result = nn.feedforward(a)

print(result)
```
